```python
def enqueue():
  n= int(input("enter the element to be inserted into the queue"))
  queue.append(n)
def dequeue():
  if len(queue)==0:
    print("queue is empty")
  else:
    print(queue[0],"is deleted from queue")
    del queue[0]
def display():
    if len(queue)==0:
      print("queue is empty")
    else:
      print("elements of queue are")
      for elements in queue:
        print(elements)
      print("rear end of the queue is",queue[-1])
      print ("front of queue is", queue[0])
queue = list()
while 1:
  print ("enter the option from below\n1-enqueue operation\n2-dequeue operation\n3-dispaly\n enter any key to exit the program")
  option = input()
  if option =='1':
    print ("enqueue operation")
    enqueue()
  elif option =='2':
    print ("dequeue operation")
    dequeue()
  elif option =='3':
    print ("display operation")
    display()
  else:
    break
```
